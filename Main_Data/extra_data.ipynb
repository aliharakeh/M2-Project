```python
df = pd.read_csv('all.csv', header=0)
```

```python
df.head()
```

```python
df[['year', 'month', 'day']] = df.date.str.split('-', expand=True)
```

```python
kadaas = df.groupby(['year', 'month', 'KADAA_ID'])
len(kadaas.groups)
```

```python
count_col = []
for n, g in kadaas:
    if n[1] != '01':
        count_col.append(len(g))
count_col = pd.Series(count_col)
count_col
```
