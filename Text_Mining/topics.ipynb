```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('all.csv', header=0)
```

```python
m = df.groupby('MOHAFAZA_AR')
```

```python
len(m)
```

```python
stopwords = []

with open('../stop-words/arabic.txt', 'r', encoding='utf-8') as f:
    stopwords += f.read().split('\n')

with open('../stop-words/english.txt', 'r', encoding='utf-8') as f:
    stopwords += f.read().split('\n')

stopwords += ['com', 'd9', 'd8', 'شو', 'أن', 'مش', 'يا', 'http', 'إلى', 'عم', 'pic', 'twitter', 'الله', 'بس', 'ly', 'شي' , 'انو',
              'انو', 'رح', 'bit', 'يلي', 'https', 'اللي', 'صباح_الخير', 'saudi24_tv', 'ليرتفع']    

# stopwords
```

```python
data = {'words': []}
index = []

for n, g in m:
    index.append(n)
    data['words'].append(" ".join(g.text))

new_df = pd.DataFrame(data=data, index=index)
new_df
```

```python
# TF-IDF Transformer
tf_Idf = TfidfVectorizer(stop_words=stopwords, ngram_range=(2, 3))

# Transform & Fit Data
tf_Idf_fit = tf_Idf.fit_transform(data['words'])

# Get Features (Columns)
feature_names = tf_Idf.get_feature_names()

# Transform the Sparse Matrix
dense = tf_Idf_fit.todense()
denselist = dense.tolist()

# Create A DataFrame
documentA_tf_idf = pd.DataFrame(denselist[0], columns=['TF-IDF'], index=feature_names)
sort = documentA_tf_idf.sort_values('TF-IDF', ascending=False)
```

```python
documentA_tf_idf = pd.DataFrame(denselist[5], columns=['TF-IDF'], index=feature_names)
sort = documentA_tf_idf.sort_values('TF-IDF', ascending=False)
sort.head(50)
```
