```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
corona = pd.read_csv('corona_lb_ratios.csv')
tweets = pd.read_csv('tweets_sentiment.csv')
```

```python
corona_months = list(corona.month.sort_values().unique())
corona_months
```

```python
tweets_months = list(tweets.month.sort_values().unique())
tweets_months
```

```python
X = corona_months
sentiment = tweets.groupby('month').mean()['sentiment']
cumlative_avg_ratio = -corona.groupby('month').mean()['cumulative_avg_ratio']
Zero = [0 for _ in X]
overall_situation = (sentiment + cumlative_avg_ratio) / 2
```

```python
plt.figure(figsize=(20, 7))
plt.plot(
    X, Zero, 'black',
    X, sentiment, 'orange',
    X, cumlative_avg_ratio, 'green',
    X, overall_situation, 'r'
)
```
