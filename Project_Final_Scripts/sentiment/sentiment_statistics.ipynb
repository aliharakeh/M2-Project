```python
import pandas as pd
import matplotlib.pyplot as plt
```

# config

```python
tweets_data = '..\\tweets_sentiment.csv'

pd.set_option('display.max_rows', 255)

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
```

# load tweets

```python
tweets = pd.read_csv(tweets_data, header=0)
tweets.head()
```

# statistics

```python
config = {
    'month': 'first',
    'day': 'first',
    'sentiment': 'mean',
    'sentiment2': 'mean'
}
days = tweets.groupby(['month', 'day']).agg(config).reset_index(drop=True)
days
```

```python
d_x = range(len(days))
d_textblob_sentiment = days.sentiment
d_flair_sentiment = days.sentiment2
```

```python
plt.plot(d_x, d_textblob_sentiment, 'orange')
```

```python
plt.plot(d_x, d_flair_sentiment, 'orange')
```

# months statistics

```python
config = {
    'month': 'first',
    'sentiment': 'mean',
    'sentiment2': 'mean',
    'text': 'count'
}
months = tweets.groupby('month').agg(config).reset_index(drop=True)
months
```

```python
m_x = months.month
m_textblob_sentiment = months.sentiment
m_flair_sentiment = months.sentiment2
tweets_count = months.text
```

```python
fig, (a, b) = plt.subplots(1, 2, figsize=(20, 10))

a.set_title('Sentiment')
a.plot(m_x, m_textblob_sentiment, 'orange')
a.grid()

b.set_title('tweets')
b.bar(m_x, tweets_count)
b.grid()
```

```python
fig1, (a1, b1) = plt.subplots(1, 2, figsize=(20, 10))

a1.set_title('Sentiment')
a1.plot(m_x, m_flair_sentiment, 'orange')
a1.grid()

b1.set_title('tweets')
b1.bar(m_x, tweets_count)
b1.grid()
```
