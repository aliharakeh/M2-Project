```python
import pandas as pd
import json
```

# Config

```python
corona_data = 'owid-covid-data.csv'
save_to = 'corona_lb_data.csv'
pd.set_option('display.max_rows', 255)
```

# load data

```python
corona_data = pd.read_csv(corona_data, header=0)
corona_data.head()
```

# columns

```python
list(corona_data.columns)
```

# get lebanon data

```python
corona_lb = corona_data[corona_data.iso_code == 'LBN'].copy().reset_index(drop=True)
```

# split date

```python
corona_lb[['year', 'month', 'day']] = corona_lb.date.str.split('-', expand=True)
```

# remove 2019 data and Jan, 2020 data (data before corona)

```python
corona_lb = corona_lb[(corona_lb.year == '2020') & (corona_lb.month > '01')]
```

# get needed columns

```python
corona_lb = corona_lb[['date', 'year', 'month', 'day', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths']]
```

# clean the memory

```python
corona_data = None
```

# save data

```python
corona_lb.to_csv(save_to, index=False)
```
