```python
import pandas as pd
```

```python
df = pd.read_csv('corona_lb_dashboard_data.csv', header=0)
```

```python
df
```

```python
df.describe()
```

```python
prev_cases = prev_death = total_cases = total_deaths = 0
first = 1

def get_data(row):
    global prev_cases, prev_death, total_cases, total_deaths, first
        
    total_cases += row.new_cases
    total_deaths += row.new_deaths
    
    data = [prev_cases, prev_death, total_cases, total_deaths]
    
    if not first:
        prev_cases = row.new_cases
        prev_death = row.new_deaths
    else:
        first = 0
    
    return pd.Series(data)
```

```python
df[['prev_cases', 'prev_deaths', 'total_cases', 'total_deaths']] = df.apply(get_data, axis=1)
```

```python
df.to_csv('corona_lb_data.csv', index=False)
```
