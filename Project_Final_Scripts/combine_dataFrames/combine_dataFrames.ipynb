```python
import os
import pandas as pd
```

```python
csv_files = []
for file in os.listdir():
    if file.endswith('.csv'):
        csv_files.append(file)

csv_files
```

```python
columns = [
    'username',
    'time',
    'text',
    'tags',
    'translated_text',
    'location',
    'latitude',
    'longitude',
    'KADAA_ID',
    'KADAA_AR',
    'KADAA_EN',
    'MOHAFAZA_ID',
    'MOHAFAZA_AR',
    'MOHAFAZA_EN'
]
```

```python
main_df = pd.DataFrame(columns=columns)
for file in csv_files:
    df = pd.read_csv(file, header=0, encoding='utf-8')[columns]
    main_df = main_df.append(df.copy())

main_df.to_csv('tweets_data.csv', index=False)
```
